```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
k1 = 2.5
d1 = 0.693147/(5)
k2 = 1000
d2 = 0.693147/(35)
```

```python
def f(s,t,arg1,arg2,arg3,arg4):
    k1,d1,k2,d2 = arg1,arg2,arg3,arg4
    m = s[0]
    p = s[1]
    
    dmdt = k1 - d1*m
    dpdt = k2*m - d2*p
    return [dmdt, dpdt]
```

```python
t = np.linspace(0,300)
s0=[0,0]

s = odeint(f,s0,t, args=tuple((k1,d1,k2,d2)) )
```

```python
plt.plot(t,s[:,0],'r--', linewidth=2.0)

plt.xlabel("t")
plt.ylabel("S[m]")
plt.legend(["[mRNA]"])
plt.show()

```

```python
plt.plot(t,s[:,1],'b-', linewidth=2.0)
plt.xlabel("t")
plt.ylabel("S[p]")
plt.legend(["[protein]"])
plt.show()
```

```python
def f2(s2,t,k1,d1,k2,d2):
    m = s2[0]
    p = s2[1]
    
    dmdt = 0
    dpdt = k2*k1/d1 - d2*p
    return [dmdt, dpdt]
t = np.linspace(0,300)
s0=[0,0]

s2 = odeint(f2,s0,t, args=tuple((k1,d1,k2,d2)) )
#term for transcription is changed
#inducer concentration is overlooked
#mRNA and protein concentration stay the same
```

```python
plt.plot(t,s[:,1],'b-', linewidth=2.0)
plt.plot(t,s2[:,1],'r-', linewidth=2.0)
plt.xlabel("t")
plt.ylabel("S[protein]")
plt.legend(["[P],[P2]"])
plt.show()
```
