```python
from scipy.optimize import fsolve
import random as r

def myFunction(z):
   u = z[0]
   v = z[1]
   

   F = np.empty((2))
   F[0] = alpha1/(1+v**b) - u
   F[1] = alpha2/(1+u**g) - v
   
   return F

s=[]
for i in range(10):
    for j in range(10):
        zGuess=[i,j]
        z=fsolve(myFunction,zGuess,full_output=True,xtol=10**-20)
        x=[np.around(z[0][0],decimals=2),np.around(z[0][1],decimals=2)]
        s.append(tuple(x))

#print(s)# s is list of tuples     
S=list(set(s)) # S is nparray of tuples
print(S)
print(np.array(S[0]))
```

```python
def J(u,v):
    j=[[-1, (-alpha1*b*v**(b-1))/(1+v**b)**2], [(-alpha2*g*u**(g-1))/(1+u**g)**2,-1]]
    #print(j)
    
    return j


x=[]
for s in S:
    x.append(J(s[0],s[1]))#apply J to each tuple in S
print(x)
```

```python
for i in x:
    v=np.linalg.eigvals(i)
    print(v)
    
#eigenvalues
```
