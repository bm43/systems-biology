```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#q 1.4
a=10
b=1000
n=2


def f(s,t,a,b,n):
    X1 = s[0]
    Y1 = s[1]
    X2 = s[2]
    Y2 = s[3]
    X3 = s[4]
    Y3 = s[5]
    
    dX1dt = a/(1+Y3**n)-X1
    dY1dt = b*X1-b*Y1
    dX2dt = a/(1+Y1**n)-X2
    dY2dt = b*X2-b*Y2
    dX3dt = a/(1+Y2**n)-X3
    dY3dt = b*X3-b*Y3
    return [dX1dt,dY1dt, dX2dt,dY2dt, dX3dt,dY3dt]


t = np.linspace(0,200,2000)
```

```python
# q 1.4 simulation + q 1.6
#q 1.6
init=[[0,0,0,0,0,0],[10**-7,10**-5,0,0,0,0],[10**-5,10**-3,0,0,0,0],[1,1000,0,0,0,0]]

for i in init:
    s = odeint(f,i,t, args=tuple((a,b,n)) )
    
    plt.plot(t,s[:,0],'r-', linewidth=2.0)
    plt.plot(t,s[:,2],'b-',linewidth=2)
    plt.plot(t,s[:,4],'g-',linewidth=2)
    plt.xlabel("t")
    plt.ylabel("mRNA")
    plt.show()
    
    plt.plot(t,s[:,1],'y-',linewidth=2)
    plt.plot(t,s[:,3],'m-',linewidth=2)
    plt.plot(t,s[:,5],'k-',linewidth=2)
    plt.xlabel("t")
    plt.ylabel("protein")
    plt.show()
```
